```python
import pandas as pd
```

```python
df = pd.read_csv('data/Country-data.csv')
```

```python
df['life_expec'].max()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('country',axis =1)  
y = df['country']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[0][0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_scaled_df = pd.DataFrame(X_scaled,columns=X.columns)
plt.figure(figsize = (20,10))       
sns.heatmap(X_scaled_df.corr(),annot = True)
```

```python
from sklearn.decomposition import PCA
```

```python
X_pca = PCA(n_components=0.9, random_state=42).fit(X_scaled)
len(X_pca.components_)
```

```python
X_pca.explained_variance_ratio_[0]
```

```python
import numpy as np
```

```python
pca = X_pca.fit_transform(X_scaled)
corrmat = np.corrcoef(pca.transpose())
sns.heatmap(corrmat, annot=True)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for i in range_n_clusters:
    kmeans = KMeans(n_clusters=i, random_state = 1)
    kmeans.fit(pca)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(pca, cluster_labels)
    print(i, silhouette_avg)
```

```python
kmeans = KMeans(n_clusters=4, random_state=1)
kmeans.fit(pca)
df['K-Means_Cluster_ID'] = kmeans.labels_
plt.figure(figsize=(8,4),dpi=200)
sns.boxplot(x='K-Means_Cluster_ID', y='gdpp', data=df)
plt.figure(figsize=(8,4),dpi=200)
sns.boxplot(x='K-Means_Cluster_ID', y='child_mort', data=df)
```

```python
K_Means_countries = df[df['K-Means_Cluster_ID'] == 1]
K_Means_countries.sort_values(by = 'life_expec')
```
